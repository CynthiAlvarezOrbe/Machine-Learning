# Parte II – Classificação binária com redes MLP e SVMs

Considere o conjunto de dados rotulados disponível no arquivo *dados_treinamento.mat*. A matriz $\mathbf{X} \in \mathbb{R}^{\mathrm{N}\times2}$ contém os atributos (*features*) dos *N* padrões existentes, enquanto o vetor $\mathbf{y} \in \mathbb{R}^{\mathrm{N}\times1}$
traz o rótulo da classe correspondente.
No caso, 
\begin{equation}
\mathrm{y}(i)=\left\{\begin{matrix}
+1, & \mathrm{se} \; \mathbf{x}(i) \in \mathrm{C_+}\\ 
-1, & \mathrm{se} \; \mathbf{x}(i) \in \mathrm{C_-}\\
\end{matrix}\right.
\end{equation}


Primeiro, vamos obter o conjunto de dados de treinameno, validação e teste.

```python
#%%%%%%%%%%%%%%%%%%%%  Primeiro, vamos importar o conjunto de dados de treinameno, validação e teste.     %%%%%%%%%%%%%%%%%%%%%%%

dadosTreinamento =  scipy.io.loadmat('dados_treinamento.mat')
dadosTeste       =  scipy.io.loadmat('dados_teste.mat')
dadosValidacao   =  scipy.io.loadmat('dados_val.mat')

xTrain  =  dadosTreinamento['X']
yTrain  =  dadosTreinamento['y']
train   =  np.concatenate((xTrain, yTrain), axis=1)

xTeste  =  dadosTeste ['Xt']
yTeste  =  dadosTeste ['yt']
teste   =  np.concatenate((xTeste , yTeste ), axis=1)

xValidacao  =  dadosValidacao['Xval']
yValidacao  =  dadosValidacao['yval']
validacao   =  np.concatenate((xValidacao, yValidacao), axis=1)

dfDadosTreinamento = pd.DataFrame(train)
dfDadosTeste       = pd.DataFrame(teste)
dfDadosValidacao   = pd.DataFrame(validacao)

names=['x_1','x_2','y']

dfDadosTreinamento.columns = names
dfDadosTeste.columns       = names
dfDadosValidacao.columns   = names

classeNegativa = dfDadosTreinamento.loc[dfDadosTreinamento['y'] == -1]
classePositiva = dfDadosTreinamento.loc[dfDadosTreinamento['y'] ==  1]
fig1 = plt.figure(figsize=(10.0, 6.0))
plt.plot(classePositiva['x_1'],classePositiva['x_2'], 'o',color='blue', label='$C_+$', markersize=3.5)
plt.plot(classeNegativa['x_1'],classeNegativa['x_2'], 'o',color='red' ,label='$C_-$', markersize=3.5)
plt.xlabel('$x_1$',fontsize='xx-large')
plt.ylabel('$x_2$',fontsize='xx-large')
plt.title('Distribuição dos dados de treinamento',fontsize='x-large')
plt.legend(loc = 0, fontsize='x-large');
plt.grid()

print(     '________________________________')
print('\033[1m' + '       Dados do problema '+ "\033[0;0m")
print('--------------------------------')
print(xTrain.shape[0], 'amostras para treinamento')
print(xValidacao.shape[0], 'amostras para validação')
print(xTeste.shape[0], 'amostras para teste')
print('________________________________')
```

A tabela acima mostra os dados de treinamento disponíveis de nosso problema. As cores estão relacionadas à classe a qual cada padrão pertence: os pontos em azul pertencem à classe $\mathrm{C_+}$,
enquanto os pontos em vermelho pertencem à classe $\mathrm{C_-}$.

Nesta atividade, vamos explorar dois métodos de classificação não-linear: uma rede neural MLP (*multilayer perceptron*) com uma única camada intermediária e uma máquina de vetores-suporte (SVM, do inglês *support vector machine*).

Vamos treinar nosso modelo utilizando o TensorFlow - Keras. Neste caso, vamos importar nosso estimador (o modelo Multi-Layer Perceptron Classifier) desta biblioteca.

```python
# Importando os módulos de keras e alguns de sklearn
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
from keras.models import load_model
from keras.layers import LeakyReLU
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

Em seguida, configuramos nossa rede com base na dimensões dos dados.

```python
#%%%%%%%%%%%%%%%%%%%%  Def de funções a serem utilizadas.     %%%%%%%%%%%%%%%%%%%%%%%

def create_model_sigmoid(activacion,numNeuronios,weight):
    
    #%%%%%     Precisamos obter as dimensões das entradas de nossos dados de entrada e teremos um nó de saída.     %%%%%
    
    numParametros = xTrain.shape[1]
    outputDimention = 1 # Binary classification

    #%%%%%%%%%%%%%%%%%%%%%%%%%%%     Definindo a rede     %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
    model = Sequential()
    #fully_connected layers
    model.add(Dense(numNeuronios, input_dim=numParametros, activation=activacion, kernel_initializer=weight))
    model.add(Dense(outputDimention, activation='sigmoid',kernel_initializer=weight))
    
    #%%%%%     Compilando o modelo     %%%%%%
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model


def create_model_than(activacion,numNeuronios,weight):
    
    #%%%%%     Precisamos obter as dimensões das entradas de nossos dados e teremos um nó de saída.     %%%%% 
    numParametros = xTrain.shape[1]
    outputDimention = 1 # Binary classification 

    #%%%%%%%%%%%%%%%%%%%%%%%%%%%     Definiendo a rede     %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
    model = Sequential()
    #fully_connected layers
    model.add(Dense(numNeuronios, input_dim=numParametros, activation=activacion, kernel_initializer=weight))
    model.add(Dense(outputDimention, activation='tanh')) 
    
    #%%%%%     Compilando o modelo      %%%%%%
    model.compile(loss='hinge', optimizer='adam', metrics=['accuracy'])
    
    return model

def plot_J(history):
    # Plotando os dados de treinamento e validação
    fig1 = plt.figure(figsize=(7.0, 4.0))
    
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.legend(['Treinamento', 'Validação'], loc=0, fontsize='large')
    
    plt.xlabel('$épocas$',fontsize='large')
    plt.ylabel('$Erro$',fontsize='large')
    plt.title('Curva de evolução do valor da função custo  $J$(.)',fontsize='large')
    pylab.show()
    return

def plot_confusion_matrix(confusion_matrix, classes, normalize = False, title='Confusion matrix', cmap=plt.cm.Blues):

    plt.imshow(confusion_matrix, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=1)
    plt.yticks(tick_marks, classes)

    fmt = '.2f'
    
    thresh = confusion_matrix.max() / 2.
    
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        
        plt.text(j, 
                 i, 
                 format(cm[i, j], fmt),
                 horizontalalignment="center", size=15,
                 color="white" if i == j else "black")#"white" if cm[i, j] > thresh else "black"

    plt.ylabel('Classe verdadeira')
    plt.xlabel('Classe estimada')
    plt.tight_layout()

```

A arquitetura da rede que construiremos terá uma camada de entrada, uma camada intermediaria e uma camada de saída. As camadas são densas (*fully-connected*).



<div align="center">
  <font size="4" color="INDIGO">[camada de entrada $_1$] <font size="4" color="red">===></font> [camada intermediária $_2$] <font size="4" color="red">===></font> [camada de saída $_3$] </font>
</div>




Onde o número de neurônios em cada camada é:

\begin{equation}
\mathbf{s_1}=2 \; \; \; \text{(número de atributos de entrada, sem contar o bias)}\\
\mathbf{s_2}= \#neurônios \; \; \; \text{(número de neurônios da capa intermédia, sem contar o bias)}\\
\mathbf{s_3}=1 \; \; \; \text{(cassificação binária)}\\
\end{equation}

### Configuração dos parâmetros da rede:

Vamos fazer um algoritmo de varredura de entre os parâmetros a seguir. Escolheremos aqueles que apresentarem melhor desempenho no conjunto de treinamento e validação de nosso problema. O algoritmo tentará encontrar o melhor configuaração para os seguintes parâmetros:

**Número de neurônios da camda intermédia**

Vamos escolher entre un cojunto de valores entre 1 e 600.
                                              

**Inicialização dos Pesos**

Os métodos de treinamento de redes neurais MLP são iterativos e de busca local, então a solução obtida depende de uma inicialização. Vamos testar diferentes opções de inicialização de peso, como:

- <font color="brown"> 'random_uniform '</font> .- distribuição uniforme
- <font color="brown"> 'glorot_normal '</font>.- distribuição aleatória uniforme [$-x$,$x$] sendo $x= \sqrt{\frac{6}{n_{inputs}+n_{outputs}}}$
- <font color="brown"> 'glorot_uniform '</font> .- distribuição aleatória uniforme [$-x$,$x$] sendo $x= \sqrt{\frac{2}{n_{inputs}+n_{outputs}}}$


**Funções de ativação**

*Camada intermediária*

Vamos testar diferentes opções de funções de ativação mais utilizadas na prática [1], como:

- Relu
- Elu
- Selu

*Camada de saída*

Vamos usar as seguintes funções de ativação na última camada, que de acordo [1], são as que apresentam melhor desempenho na camada de saída e são as que melhor se adaptam a nosso problema. 
- Sigmóide [0,1]
- Tangente hiperbólica [-1,1]


**Otimizador**

Nosso conjunto de dados não era particularmente grande (1000 amostras). Vamos utilizar uma técnica mini-batch com um batch de 20% do conjunto de dados (200 amostras). O otimizador que vai ser usado é  <font color="brown">'Adam'</font>: Adaptive Moment Estimation. Este otimizador usa técnicas de aprendizado adaptativo que podem superar outras técnicas adaptativas, com a vantagem de que não precisamos escolher o *learning rate* ($\alpha$).


**Função Custo**

Vamos aplicar a função de entropia cruzada para o problema de classificação binária: 
- <font color="brown">'binary_crossentropy'</font> no caso da função sigmoide na camada de saída e, 
-  <font color="brown">'hinge'</font> no caso da função tanh


**Métricas**

As métricas que vamos utilizar para avaliar nosso modelo são: 
- Acurácia : Keras 
- Matriz de Confusão : scikit-learn


Vamos exibir duas tabelas com os parâmetros ótimos encontrados, avaliando a acurácia para 300 épocas, para cada função de ativação na camada de saída.

```python
#%%%%%%%%%%%%%% Procura os melhores parâmetros com a Função de Ativação Tangente Hiperbolica da camada de saída %%%%%%%%%%%%%%%%%%
epocas = 300
batch = 200 # Batch para atualizacão dos pesos 20% do conjunto 

activation2 = ['relu', 'elu','selu'] 
neur=[1,50,100,200]
pesos= ['random_uniform','glorot_normal','glorot_uniform']
acc=0

for i in neur:
    for k in activation2:
        for h in pesos:
            model=create_model_than(k,i,h);
            model.fit(xTrain, yTrain.ravel(), batch_size=batch, epochs=epocas, verbose=0, validation_data=(xValidacao, yValidacao.ravel()));
            score = model.evaluate(xValidacao, yValidacao, verbose=2);
            if score[1]> acc:
                n_n=i
                f_a=k
                initi=h
                acc=score[1]
                    
model=create_model_than(f_a,n_n,initi)
model.fit(xTrain, yTrain.ravel(), batch_size=batch, epochs=epocas, verbose=0, validation_data=(xValidacao, yValidacao.ravel()));
score = model.evaluate(xTrain, yTrain, verbose=2)

print(' _______________________________________________________________________________________')
print('\033[1m' + '          Função de ativação da camada de saída :  tangente hiperbolica '+ "\033[0;0m")
print(' _______________________________________________________________________________________')
print('   PARâMETROS ÓTIMOS  avaliando a acurácia :   acc_test:  %.2f%%  e   acc_val:  %.2f%%    ' %(score[1]*100,acc*100))
print(' ---------------------------------------------------------------------------------------')
print(' n. neurônios camada intermediária : [1,50,100,200]                     ----> ',n_n)
print(' função de ativação camada intermediária : [ relu, elu, selu ]          ----> ',f_a)
print(' inicialização pesos: [random_uniform,glorot_normal,glorot_uniform]  ---->',initi)
print(' _______________________________________________________________________________________')
```

```python
#%%%%%%%%%%%%%% Procura melhores parâmetros com Função de activação da camada de saída :  sigmóide %%%%%%%%%%%%%%%%%%
epocas = 300
batch = 200 # Batch para actualizacão dos pesos 20% do conjunto 

yTrain[yTrain == -1] = 0
yValidacao[yValidacao == -1] = 0
yTeste[yTeste == -1] = 0

activation2 = ['relu', 'elu','selu'] 
neur=[1,50,100,200,400,600,800]
pesos= ['random_uniform','glorot_normal','glorot_uniform']
acc=0

for i in neur:
    for k in activation2:
        for h in pesos:
            model=create_model_sigmoid(k,i,h);
            model.fit(xTrain, yTrain.ravel(), batch_size=batch, epochs=epocas, verbose=0, validation_data=(xValidacao, yValidacao.ravel()));
            score = model.evaluate(xValidacao, yValidacao, verbose=2);
            if score[1]> acc:
                n_n=i
                f_a=k
                initi=h
                acc=score[1]
                    
model=create_model_sigmoid(f_a,n_n,initi)
model.fit(xTrain, yTrain.ravel(), batch_size=batch, epochs=epocas, verbose=0, validation_data=(xValidacao, yValidacao.ravel()));
score = model.evaluate(xTrain, yTrain, verbose=2)

print(' _______________________________________________________________________________________')
print('\033[1m' +'              Função de activação da camada de saída :  sigmóide '+ "\033[0;0m")
print(' _______________________________________________________________________________________')
print('    PARAMETROS OTIMOS  evaluando acurácia :   acc_test:  %.2f%%  e   acc_val:  %.2f%%    ' %(score[1]*100,acc*100))
print(' ---------------------------------------------------------------------------------------')
print(' n. neurônios camada intermédia : [1,50,100,200,400,600,800]         ----> ',n_n)
print(' função de ativação camada intermédia : [ relu, elu, selu ]          ----> ',f_a)
print(' inicialização pesos: [random_uniform,glorot_normal,glorot_uniform]  ---->',initi)
print(' _______________________________________________________________________________________')
```

Comparando as duas tabelas, podemos dizer que a função de ativação sigmóide apresentou melhor desempenho na camada de saída, apresentando um valor de acurácia no conjunto de validação e teste superiores (>80%) à função tangente hiperbólica. Nos próximos passo vamos treinar a rede para 1000 épocas e olhar nossa função de custo em cada época. 

```python
model=create_model_sigmoid('relu',200,'glorot_normal')
epocas=2000
batch=200

history = model.fit(xTrain, yTrain.ravel(), batch_size=batch, epochs=epocas, verbose=0, validation_data=(xValidacao, yValidacao.ravel()));
score = model.evaluate(xTrain, yTrain, verbose=2)
print('J_Treinamento:  %.3f,   acc: %.2f%%' % (score[0], score[1]*100))
score = model.evaluate(xValidacao, yValidacao, verbose=2)
print('J_Validação:   %.3f,    acc: %.2f%%' % (score[0], score[1]*100))
plot_J(history);
```

Na figura acima, podemos ver a diminução da função custo por época o que significa que o algoritmo está funcionando bem. Agora vamos a empregar a técnica de regularização para evitar o overfitting no nosso modelo (early stooping + validação cruzada (hold out)).

```python
model=create_model_sigmoid('relu',200,'glorot_normal')
# Evitar o Overfit da rede MLP com Early Stopping
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=30)
mc = ModelCheckpoint('best_model.h5', monitor='val_acc', mode='max', verbose=0, save_best_only=True)
history = model.fit(xTrain, yTrain.ravel(), batch_size=batch, epochs=epocas, verbose=0, validation_data=(xValidacao, yValidacao.ravel()),callbacks=[es, mc]);
score = model.evaluate(xTrain, yTrain, verbose=2)
print('J_Treinamento: %.3f,   acc: %.2f%%' % (score[0], score[1]*100))
score = model.evaluate(xValidacao, yValidacao, verbose=2)
print('J_Validação:   %.3f,   acc: %.2f%%' % (score[0], score[1]*100))
plot_J(history);
```

|  paciência=30 |   sem regularizaçao  |      Regularização |
|----------------------|-------|-------|
|   $J_{Treinamento}$ |  0,258 | 0,278| 
|   $J_{Validação}$ |  **0,324**| **0,325**| 
|   épocas |  1000| 387| 

O critério utilizado para interromper o treinamento e evitar o *overfitting* foi a paciência. Se aumentarmos o número de épocas (387 a 2000) o $J_{val}$ se mantem em um valor $\approx $ de 0.324 constante, enquanto o $J_{train}$ continua diminuindo ao aumentar o número de épocas.

Ao diminuir o número de épocas de treinamento, podemos reduzir o custo computacional de treinamento e evitar um sobreajuste de nosso modelo, conseguindo que o modelo possa generalizar bem. Então vamos utilizar este modelo com regularização.

```python
yTeste[yTeste == -1] = 0
n_hidden = [2, 5, 10, 15, 25, 50, 100, 200, 500, 800, 1000,2000]
epocas=325
batch=200

print(' __________________________________________________________________________')
print('\033[1m' +'  Número de Neurônios |              Acurácia   '+ "\033[0;0m")
print(' __________________________________________________________________________')

for n in n_hidden:
    
    model=create_model_sigmoid('relu',n,'glorot_normal')
    history = model.fit(xTrain, yTrain.ravel(), batch_size=batch, epochs=epocas, verbose=0, validation_data=(xValidacao, yValidacao.ravel()));
    score = model.evaluate(xTeste, yTeste, verbose=2)
    scoret = model.evaluate(xTrain, yTrain, verbose=2)
    scorev = model.evaluate(xValidacao, yValidacao, verbose=2)
    print('        %d       acc_train: %.2f%%   acc_val: %.2f%% ' % (n, scoret[1]*100,scorev[1]*100))
    
print(' __________________________________________________________________________')
    
```

**(e)** Faça, agora, o projeto de uma SVM para a mesma tarefa de classificação.
Mostre as regiões de decisão definidas pela SVM no espaço original dos dados
e os vetores-suporte identificados.

O Support Vector Machine busca gerar um hiperplano para separar os dados e busca maximizar as margens, ou seja, a distância entre o hiperplano e os padrões de cada classe mais próximos, como pode ser visto na figura a seguir.  

Para implementar um classificador através de Support Vector Machine, escolheu-se a biblioteca SciKit Learning para isso, através do sklearn.svm.SVC. O SVC do SciKit Learning possui diversos parâmetros e atributos. Pode-se passar para a função de treinamento, o valor de __C__ , escolher o tipo de kernel, entre outros parâmetros. Caso esses dados não sejam passados para o SVC, o SVC acaba utilizando valores padrão (default).

<img src="files/svm_margem.png">

Para utilizar o SVM para este problema de classificação, em que os dados não são linearmente separáveis e a fronteira de decisão não é linear, deve-se utilizar um mapeamento não-linear, que leva estes dados para outro espaço de maior dimensão. Desta maneira, torna-se mais fácil separá-los. Utiliza-se então as funções de kernel para realizar esta operação. O svm.SVC do SciKit Learning dispõe de algumas possibilidades de escolha do *kernel*, como pode ser visto a seguir:

- kernel **'linear'**: O kernel Linear é utilizado quando os dados são linearmente separáveis, ou seja, podem ser separados por uma única linha. 
- kernel **'poly'**: Neste kernel, a função representa a similaridade de vetores em um espaço de características de polinômios das variáveis originais.
- kernel **'rbf'**: Este kernel tem a forma de uma função de base radial. É bastante popular e utilizada.
- kernel **'sigmoid'**: Este kernel vem da área de Redes Neurais e possui uma função bipolar, muito utilizada como função de ativação em MLP.

Escolheu-se o kernel Radial Basis Function (RBF) porque a fronteira é não-linear e com esse tipo de fronteira, este tipo de kernel costuma ter um bom desempenho. 

 **C** é um parâmetro de regularização que diz o quanto se quer evitar classificar errado. Para valores grande de __C__, a otimização escolherá hiperplanos com uma margem menor se isso melhorar a classificação. Para valores pequenos de __C__, a otimização escolherá hiperplanos que terão margens maiores, mesmo que isso gere mais classificações erradas.

Para escolher o valor de __C__, utilizou-se o conjunto de dados de treinamento e validação. Esse parâmetro não foi passado porque o default do SVM já é esse tipo de kernel. Treinou-se vários modelos com os dados de treinamento para vários valores de C. Os dados de validação foram usados e o que gerou a maior acurácia (acurácia é o complemento do erro, então maior acurácia significa menor erro) foi escolhido como pode ser visto a seguir. Os valores de **C** escolhidos para comparação foram: 0.5, 1.0, 1.5, 3.0, 5.0 e 10.0.

Outro parâmetro importante é o __gamma__ . A função do kernel rbf é uma função de similaridade entre dados de treinamento e precisa ser ajustada por um fator __gamma__. Este parâmetro pode ser visto como o inverso do raio de influência das amostras selecionadas pelo modelo como vetores de suporte. Se __gamma__ é muito grande, o raio da área de influência incluirá apenas o próprio vetor, enquanto que se __gamma__ é muito pequeno, o modelo fica muito restrito e não consegue capturar a complexidade dos dados, o raio de influência incluirá quase todo o conjunto. O __gamma__ foi setado como escalar (*'scale'*) para levar a variância dos dados em consideradação.

```python
data_train_x = xTrain.copy()
data_train_y = yTrain.copy()
data_val_x =  xValidacao.copy()
data_val_y = yValidacao.copy()
data_test_x = xTeste.copy()
data_test_y = yTeste.copy()
```

```python
#Vetor que guardará todos os valores de acurácia para serem plotadas
scores_acc = np.zeros([10,1])

#Para vários valores de C, o modelo é treinado e acurácia para os dados de validação são obtidos.
for C in range (1,11):
    clf = svm.SVC(C=C, gamma='scale');
    clf.fit(xTrain, yTrain.ravel());
    previsao = clf.predict(xValidacao);
    print('Resultado para C=%d :' % (C))
    print(accuracy_score(yValidacao, previsao))
    scores_acc[C-1] = accuracy_score(yValidacao, previsao)

#Plota o gráfico de acurácia para cada C
plt.plot(scores_acc)
plt.xlabel(['Valores de C'])
plt.ylabel(['Valores de Acurácia'])
plt.title('Acurácia para cada valor de C')
plt.ylim(bottom=0.7)
plt.show()
```

Como é possível observar nos resultados obtidos acima, a acurácia tem seus valores mais altos quando C se aproxima de 10. Entretanto, entre os mínimos e os máximos valores de C, tem-se apenas 3% de diferença. Utilizamos o modelo com C=10. Como dito anteriormente, escolhemos o kernel rbf.

```python
#Definição do modelo e parâmetros deste modelo
clf = svm.SVC(C=10, gamma='scale');

#Treinamento dos dados
clf.fit(xTrain, yTrain.ravel());

#Previsão com o Conjunto de Validação
previsao = clf.predict(xValidacao);
#Apresentação do desempenho do modelo em relação aos dados de validação
print(classification_report(yValidacao.ravel(), previsao))

#Parâmetros do gráfico
scatter_kwargs = {'s': 10, 'edgecolor': None, 'alpha': 1, 'label':'classe -1'}
scatter_highlight_kwargs = {'s': 15, 'label': 'Vetores de Suporte', 'alpha': 0.7}

#Utilizou-se a função do mlxtend 'plot_decision_regions'
plot_decision_regions(xTrain, yTrain.ravel(), clf=clf, legend=4, scatter_kwargs=scatter_kwargs,scatter_highlight_kwargs=scatter_highlight_kwargs, X_highlight=clf.support_vectors_)
plt.ylim(-2.4, 2.4)
plt.xlim(-2.5, 2.5)
plt.title('Regiões de Decisão e Vetores de Suporte')
```

Como pode ser visto anteriormente, a fronteira obtida pelo classificador é não-linear e conseguiu separar bem as duas classes. A acurácia obtida foi de 86%, significando que a taxa de erro foi de 14%. Outras métricas como precisão, recall, f1-score, entre outros, também apresentaram bons resultados. Os resultados obtidos se assemelharam aos obtidos por MLP com uma camada intermediária. A seguir, pode-se conferir a Matriz de Confusão. Observa-se que de 1000 padrões de entrada, o classificador acertou a classe de 864.

```python
#Plota a matriz de confusão
print("Correctly predicted labels:", np.sum(yValidacao.ravel() == previsao),' of:',len(yValidacao), 'acc_teste: %.2f%%'% (np.sum(yValidacao.ravel() == previsao)*100/len(yValidacao)))
classes=['1', '-1']
cm=confusion_matrix(yValidacao, previsao)
plot_confusion_matrix(cm, classes, title = "Matriz de Confusão")
```
