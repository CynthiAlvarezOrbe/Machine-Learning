## Predição de séries temporais


As técnicas de análise e predição de séries temporais são indispensáveis às ciências e engenharias [1]. Em qualquer ramo de atuação profissional, percebe-se um crescimento acentuado na
demanda por predições e detecção de tendências.

As séries temporais são conjuntos de dados que representam o comportamento de uma ou mais variáveis
no decorrer do tempo [1]. São duas as principais atividades relacionadas à
séries temporais: análise e predição. Na análise,
deseja-se descrever o comportamento da série temporal, identificar nuâncias e fatores de influência. Já na
predição o objetivo é obter os valores futuros para a série com base nos valores já conhecidos [1]. 

A fim de se prever o valor futuro de uma série de medidas de uma determinada grandeza,
um procedimento típico consiste em construir um modelo matemático de estimação
baseado na hipótese de que os valores passados da própria série podem explicar o seu
comportamento futuro.

A modelagem de uma série temporal que desejamos prever o valor um passo à frente é:

Seja $ x (n)$ o valor da série temporal no instante (discreto) $n$ . Então, o modelo
construído deve realizar um mapeamento do vetor de entradas $\textbf{x(n)} \in \mathbb{R}^{K+1}$
, o qual é
formado por um subconjunto de amostras passadas, $i.e,$

\begin{equation} 
\mathbf{ x(n) } = [x(n-1) .... x(n-K)]^T
\end{equation}

para uma saída $\widehat{y}(n)$ , que representa uma estimativa do valor futuro da série $x(n)$.

## Preditor Linear

Neste exercício, vamos trabalhar com a famosa série histórica de medidas do número de manchas solares $\textit{(sunspot)}$. No caso, dispomos das leituras mensais desde 1749 a 2014, totalizando 3458 amostras. Além disso, o modelo que será empregado na
previsão corresponde a um preditor linear, tal que:

\begin{equation} 
\widehat{y}(n) = \mathbf{w^Tx(n)}+w_0
\end{equation}

Portanto é apresentado uma discussão dos resultados obtidos no método proposto e seu desempenho.

$\textbf{Procedimento de construção do modelo de previsão:}$

A nossa série temporal tem $K=20$ variavéis de entradas, as quais correspondem a 20 atrasos consecutivos, na ordem do mais recente ao mais antigo. 

\begin{equation} 
\mathbf{x(n) } = [x_{n-1}\; x_{n-2}\; .... \;x_{n-19}\; x_{n-20}]^T \\
\end{equation}

E o modelo do preditor linear esta dado por:

\begin{equation} 
\mathbf{\widehat{y}(n)} = w_o + \mathbf{w^Tx(n)}
\end{equation}

A nossa solução do preditor pelo método de quadrados mínimos e equações normais com $K+1<N$ e de posto completo é: 

\begin{equation}
\mathbf{W } = (\mathbf{ \Phi }^T\mathbf{ \Phi })^{-1}\mathbf{ \Phi }^T\mathbf{ y }\\
\mathbf{W} = \mathbf{w_0+w}
\end{equation}

Onde a matriz $\mathbf{\Phi} \in \mathbb{R}^{N\mathbf{x}(K+1)}$ com os $N$ pares de dados correspondentes até dezembro de 2009 é:

\begin{equation} 
\mathbf{ \Phi } = \begin{bmatrix} { 1\\ .\\.\\ 1 }\end{bmatrix} +\begin{bmatrix} \mathbf{ x(0) }^T\\ .\\.\\ \mathbf{ x(N-1)^T }\end{bmatrix}\\
\end{equation}

```python
#variaveis de entrada
K=20
# Função para converter a array de valores en uma matrix dataset 
def create_dataset(dataset, K=20):
    dataX, dataY = [], []
    for i in range(len(dataset)-K-1):
        x = [1]
        a = dataset[i:(i+K)]
        x.extend(a)
        dataX.append(x)
        dataY.append(dataset[i + K])
    return np.array(dataX), np.array(dataY)      
```

```python
#Função para calcular a saída y_chapeú 

def y_chapeu(dataX,w):
    y_chapeu=([])
    a=(dataX.shape)
    n=np.array(a)[0]
    for i in range(n):
        v=np.matmul(dataX[i],w)
        y_chapeu.append(v) 
    return np.array(y_chapeu)  
```

## (b) Agora, vamos considerar a aplicação da estratégia de seleção de variáveis denominada wrapper, seguindo a abordagem backward elimination. Ademais, a fim de medirmos de forma mais robusta o desempenho de cada subconjunto de variáveis de entrada, vamos utilizar um esquema de validação cruzada do tipo k-fold. Por fim, vamos também considerar a técnica ridge regression para a regularização do modelo.  

O $wrapper$ percorrerá, com a estratégia de $Backward\; elimination$, parte do conjunto
completo de variáveis de entrada para identificar o subconjunto "ótimo" de variavéis, utilizando o desempenho médio alcançado pelo preditor como indicativo da qualidade. Enquanto, se utiliza uma estratégia para validação cruzada ($k-folders$) a fim de se ter uma estimativa suficientemente precisa do erro de generalização. 

Vamos dividir o conjunto de amostras disponíveis para treinamento em 10 pastas, de forma que toda amostra disponível vai aparecer 9 vezes no conjunto de treinamento e 1 vez no conjunto de validação. Por fim vamos introduzir um coeficiente de regularização $\lambda\in$ [0.00001, 0.0001, 0.003, 0.01 ,0.3 , 0 , 1, 10, 100,20000, 1000000 ,1000000, 1000000] para melhorar o processo de generalização do modelo (evitar o overfitting), que na solução fechada da pseudo-inversa tem o efeito colateral fazendo que o termo $(\mathbf{ \Phi }^T\mathbf{ \Phi }+\lambda \mathbf{I'})^{-1}$ sempre exista mesmo que a matriz  $\mathbf{ \Phi }$ não possua posto completo.

A nossa solução do preditor pelo método de quadrados mínimos e equações normais com $K+1<N$ é: 

\begin{equation}
\mathbf{W } = (\mathbf{ \Phi }^T\mathbf{ \Phi }+\lambda \mathbf{I'})^{-1}\mathbf{ \Phi }^T\mathbf{ y }\\
\mathbf{W} = \mathbf{w_0+w}
\end{equation}

```python
#Utilizar só o data de treinamento
X_treinamento, Y_treinamento=create_dataset(dataset)
lambida=[0.00001, 0.0001, 0.003, 0.01 ,0.3 , 0 , 1, 10, 100,20000, 1000000 ,1000000, 1000000]
```

```python
#função para dividir em 10 folders
def k_fold(lambida,X,dfY):
    
    k_folders=10
    dados=round(len(X)/k_folders)

    treinamentoScorelambida=[]
    for lam in range(len(lambida)-1):  
        treinamentoScore=[]
        for index in range(k_folders):
            teinamentoX=X.drop(X.index[index*dados:(index+1)*dados],axis=0)
            teinamentoY=dfY.drop(dfY.index[index*dados:(index+1)*dados],axis=0)
            validacaoX=X[index*dados:(index+1)*dados].values
            validacaoY=dfY[index*dados:(index+1)*dados].values        
        #Obter o vetor de parámetros
            I_prima=np.identity(len(X.columns))
            I_prima[0,0]=0
            w1=np.linalg.inv(np.matmul(np.transpose(teinamentoX), teinamentoX)+(lambida[lam]*I_prima))
            #w1=np.linalg.inv(np.matmul(np.transpose(teinamentoX), teinamentoX)+(0*I_prima))
            w2=np.matmul(w1, np.transpose(teinamentoX))
            w=np.matmul(w2,teinamentoY)
            validacaoY_chapeu=y_chapeu(validacaoX,w)
            TS=sqrt(mean_squared_error(validacaoY,validacaoY_chapeu))
            treinamentoScore.extend([TS])
        treinamentoScorekFolder=np.mean(treinamentoScore)
        treinamentoScorelambida.extend([treinamentoScorekFolder])
    lambidaotimo=lambida[np.argmin(treinamentoScorelambida)]
    RMSEminkfolder=treinamentoScorelambida[np.argmin(treinamentoScorelambida)]
    return lambidaotimo, RMSEminkfolder
```

```python
def eliminacao_variable(x_n,y_n,parametros):
    k=len(x_n.columns)
    lambidaOtimo=[]
    RMSE_lambidaOtimo=[]

    for x_i in parametros:
        X=x_n.drop([x_i],axis=1)
        lambida_RMSE=k_fold(lambida,X,y_n)

        lambidaOtimo.extend([lambida_RMSE[0]])
        RMSE_lambidaOtimo.extend([lambida_RMSE[1]])
    
    RMSE_lambidaOtimo
    lambidaotimo=lambidaOtimo[np.argmin(RMSE_lambidaOtimo)]
    RMSElambidaotimo=RMSE_lambidaOtimo[np.argmin(RMSE_lambidaOtimo)]
    x_i_eliminado=parametros[np.argmin(RMSE_lambidaOtimo)]
    return x_i_eliminado, RMSElambidaotimo,  lambidaotimo
```

```python
df = pd.DataFrame(X_treinamento)
dfY = pd.DataFrame(Y_treinamento)
#trabalhar com k=10 então eliminamos a última fila
X_n=df.drop(len(df)-1)
Y_n=dfY.drop(len(df)-1)
#X_n=df
#Y_n=dfY
RMSE=[]
Lambida=[]
lista_def=[]
lista=list(range(K+1))[1:K+1]
valores=list(range(K+2))
valores.reverse()
lista_def += [valores[1:21]]
u=k_fold(lambida,X_n,Y_n)
RMSE.extend([u[1]])
Lambida.extend([u[0]])

for z in range(K-1):
    h=z
    elimadoX=eliminacao_variable(X_n,Y_n,lista)
    RMSE.extend([elimadoX[1]])
    Lambida.extend([elimadoX[2]])
    X_n=X_n.drop(elimadoX[0],axis=1)
    lista.remove(elimadoX[0])
    variaveisSelecionados = [valores[m] for m in lista]
    lista_def += [variaveisSelecionados]
```

```python
melhoresVariaveis=lista_def[np.argmin(RMSE)]
melhoresVariaveis.reverse()
Lambida[np.argmin(RMSE)]
numeroVariaveis=(len(melhoresVariaveis))
display(Math('\mathrm{conjunto\;“ótimo”\;de }\; x_i\; \mathrm{atrasos\; identificado\; pelo}\;\mathit{wrapper}\;\mathrm{onde}\; i\; \in:') ,melhoresVariaveis)
print('Com número de variavéis igual a : %.00f' % (numeroVariaveis))
display(Math('\mathrm{Para\;um\; \lambda= }'),Lambida[np.argmin(RMSE)])
print('Com um desempenho médio do preditor linear de: %.10f RMSE' % (RMSE[np.argmin(RMSE)]))
```

O $wrapper$ escolheu um subconjunto ótimo de 12 variáveis (entre o conjunto de 20 atrasos mais recentes) com o melhor desempenho médio do preditor linear e o termo de regularização.

```python
#função para dividir em 10 folders
def k_folder(lambida,X,dfY):
    
    k_folders=10
    dados=round(len(X)/k_folders)
    vectorw=[]
    treinamentoScorelambida=[]
    #for lam in range(len(lambida)-1):  
    treinamentoScore=[]
    for index in range(k_folders):
        teinamentoX=X.drop(X.index[index*dados:(index+1)*dados],axis=0)
        teinamentoY=dfY.drop(dfY.index[index*dados:(index+1)*dados],axis=0)
        validacaoX=X[index*dados:(index+1)*dados].values
        validacaoY=dfY[index*dados:(index+1)*dados].values        
        #Obter o vetor de parámetros
        I_prima=np.identity(len(X.columns))
        I_prima[0,0]=0
        w1=np.linalg.inv(np.matmul(np.transpose(teinamentoX), teinamentoX)+(lambida*I_prima))
        w2=np.matmul(w1, np.transpose(teinamentoX))
        w=np.matmul(w2,teinamentoY)
        validacaoY_chapeu=y_chapeu(validacaoX,w)
        TS=sqrt(mean_squared_error(validacaoY,validacaoY_chapeu))
        treinamentoScore.extend([TS])
        vectorw.extend(w)
    treinamentoScorekFolder=np.mean(treinamentoScore)
    return treinamentoScorekFolder,np.array(vectorw),treinamentoScore
```

```python
X_treinamento, Y_treinamento=create_dataset(dataset)
df = pd.DataFrame(X_treinamento)
dfY = pd.DataFrame(Y_treinamento)
#trabalhar com k=10 então eliminamos a última fila
Y_n=dfY.drop(len(df)-1)
x_n=df.drop(len(df)-1)
ve=[2,4,6,7,8,9,11,14]
for x_i in ve:
    x_n=x_n.drop([x_i],axis=1)
lambida=[0 , 0.0000001, 0.00001, 0.001, 0.01 ,0.1 ,1, 10,100,500,1000,5000,10000,20000,25000,30000,40000,50000]
x_n
erro=[]
w_2=[]
for x_i in range(len(lambida)):
    a=k_folder(lambida[x_i],x_n,Y_n)[0]
    b=k_folder(lambida[x_i],x_n,Y_n)[1]
    
    erro.extend([a])
    w_2.extend([b])
erro1=erro
plt.figure(figsize=(18.0, 5.0))
plt.plot(erro,color='orange',linewidth=2)
z=np.array(erro)
plt.plot(np.argmin(z),min(erro),'b-o',label='$RMSE_{min}$',color='green', markersize=10)
plt.xlabel('$\lambda$',fontsize='x-large')
plt.ylabel('[RMSE]',fontsize='x-large')
plt.title('Desempenho médio do preditor linear com regularização no conjunto “ótimo” de atrasos identificado pelo $wrapper$',fontsize='x-large')
plt.yticks(fontsize='large')
plt.legend(loc = 0, fontsize='xx-large')
plt.xticks(np.arange(0, 18, step=1),('0','1e-07','0.00001','0.001','0.01','0.1' ,'1', '10','100','500','1000','5000','10000','20000','25000','30000','40000','50000'),fontsize='large');
plt.grid(color='grey', linestyle='-', linewidth=0.3)
display(Math('\mathrm{O\;RMSE_{min}\;do\;sub-conjunto\;de\;12\;variavéis\;identificado\;pelo}\;\mathit{wrapper}\;\mathrm{é}:') ,min(erro))
display(Math('\mathrm{Para\;um\; \lambda= }'),20000)
```

No gráfico anterior são treinados as 12 variavéis escolhidas pelo $wrapper$ utilizando $10\; folders$ para varios valores de $\lambda$ diferentes, escolhendo as amostras suficientes para fazer a predição. O hiperparâmetro $\lambda$ reduz a possibilidade de ocorrer $overfitting$ e também  mantem os pesos $w$'s do modelo o menor possível. Podemos olhar que o hiperparâmetro $\lambda$=20000 melhora o desempenho médio do preditor com relação a $\lambda$=0 em $\approx $ 0,001, melhorando a capacidade de generalização do preditor linear. Além disso, podemos também olhar que conforme o $\lambda\rightarrow\infty$ , o desempenho do preditor linear vai piorando isto é devido ao fato que os $w_i$'s acabam muito próximos de zero. $\mathbf{CONCLUSÕES:}$

$\circ $ Neste excercicío, aplicamos o modelo de um Preditor Linear usando uma equação direta de "forma fechada" que calcula diretamente os parâmetros do modelo que melhor se ajustam ao modelo para o conjunto de treinamento (ou seja, os parâmetros do modelo que minimizam a função de custo sobre o conjunto de treinamento).

$\circ $ É interessante mencionar que o uso da Equação Normal neste caso é atrativo para treinar o modelo de Regressão Linear pois essa da uma solução em forma fechada e não precisa iterar como se faría se utilizássemos outros métodos como "Gradient Descent". Contudo, se o número de coeficientes de treinamento (ou analogamente, o número de atributos $K$) não fosse K=20 e sim algo muito maior, este método teria desvantagem em termos de custo de recursos computacionais $(K^3)$ devido o calculo da pseudo-inversa da matrix $(\mathbf{ \Phi }^T\mathbf{ \Phi })^{-1}$ tornando-o já não atrativo.

$\circ $ Foram aplicados critérios de seleção das melhores entradas para a particularização do modelo com o fim de minimizar a dimensionalidade do problema. As duas estrategias de seleção comparadas foram filtro e wrapper, sendo o wrapper aquele que teve maior potencial para identificar o sub-conjunto "ótimo" de variavéis, porém mais custoso computacionalmente.

$\circ $  E finalmente se analiza o uso de $k-folder$ e $ridge$ $regression$ como estratégias que auxilia na obtenção de um preditor melhorando a capacidade de generalização e o desempenho do modelo.
